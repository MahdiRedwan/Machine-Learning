```python
import numpy as np
```

```python
# Parameters
num_images = 5000  # Number of images
image_height = 28  # Height of each image
image_width = 28   # Width of each image

# Create an array of shape (5000, 28, 28) filled with random values between 0 and 1
handsignX2 = np.random.rand(num_images, image_height, image_width)
```

```python
# Save the data to handsignX.npy
np.save('handsignX2.npy', handsignX2)  # This will save the file in the current working directory
```

```python
# Load the data back from the .npy file
loaded_handsignX2 = np.load('handsignX2.npy')

# Print the shape and some sample data to verify
print("Loaded Data Shape:", loaded_handsignX2.shape)  # Should be (5000, 28, 28)
print("Sample Data from Loaded File:", loaded_handsignX2[0])  # Display the first image
```
